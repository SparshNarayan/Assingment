```python
import numpy as np
```

```python
aravali = np.array(np.random.randint(10, 40, (8,2)))
print(f"The defined array is \n{aravali}")
print(f"The shape of the array is {aravali.shape}")
```

```python
nilgiri = np.transpose(aravali)
print(f"The transposed array is \n {nilgiri}")
print(f"The shape of transposed array is {nilgiri.shape}")
```

```python
new_3d_array = np.array(np.random.randint(5,50,(3,5,2)))
print(new_3d_array)
print(f"The shape of array is {new_3d_array.shape}")
```

```python
transposed_array = np.transpose(new_3d_array)
print(transposed_array)
print(f"The shape of the array is {transposed_array.shape}")
```

```python
#New 2d array
array_2d = np.array(np.random.randint(1,10,(2,10)))
print(array_2d)
print(array_2d.shape)
```

```python
new_shape = (2,5,2)
reshaped_array = np.reshape(array_2d, new_shape)
print(reshaped_array)
print(reshaped_array.shape)
```

```python
new_shape = (5,2,2)
reshaped_array = np.reshape(array_2d, new_shape)
print(reshaped_array)
print(reshaped_array.shape)
```
