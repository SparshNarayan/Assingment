```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_data.csv')
pd.set_option('display.max_columns' , None)
df.sample(5)
```

```python
df.info()
# budget , popularity , runtime , revenue are the numeric columns
```

```python
df.isnull().mean()*100
# Belons to Collection has the most null values in it 79.8%
```

```python
df.shape
```

```python
plt.scatter(df['budget'] , df['revenue'])
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget Vs Revenue')
plt.show()
# Almost all the High Budget Movies have High Revenue
```

```python
sns.boxplot([df['budget'] , df['revenue']])
# Revenue has Very high Outliers
```

```python
plt.hist(df['runtime'] , bins = 30)
plt.xlabel('Budget')
plt.ylabel('Number of Movies')
plt.title('Budget Distribution')
# Maximum Movies are of 90 to 110 runtime
```

```python
df['original_language'].value_counts().plot(kind = 'bar')
# Maximum Movies are of English Language and then French
```

```python
plt.hist(df['popularity'] , bins = 30)
plt.show()
# maximum movies are less popular only few movies are popular
```

```python
plt.hist(df['budget'] , bins = 30)
plt.show()
# Maximum movies are of low Budget only few movies are of high Budget
```

```python
plt.scatter(df['budget'] , df['popularity'])
plt.xlabel('Budget')
plt.ylabel('Popularity')
plt.title('Budget Vs Popularity')
plt.show()
# High Budget Movies are not popular 
```

```python
df.sort_values(by ='revenue' , ascending = False).head(10)
# Top 10 Highest Revenue Movies
```

```python
plt.bar(df.sort_values(by ='revenue' , ascending = False).head(10)['title'] , df.sort_values(by ='revenue' , ascending = False).head(10)['revenue'])
plt.xticks(rotation = 90)
plt.title('Top 10 Highest Revenue Movies')
plt.show()
```
