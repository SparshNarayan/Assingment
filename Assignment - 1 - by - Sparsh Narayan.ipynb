```python
import ast

def get_scorecard(match_index):

    match = df.iloc[match_index]
    innings_data = ast.literal_eval(match['innings'])

    scorecards = []

    for inn in innings_data:

        innings_name = list(inn.keys())[0]
        innings_info = inn[innings_name]

        deliveries = innings_info['deliveries']

        batsman_runs = {}
        bowler_wickets = {}

        for delivery in deliveries:

            ball = list(delivery.values())[0]

            batsman = ball['batsman']
            bowler = ball['bowler']
            runs = ball['runs']['batsman']

            batsman_runs[batsman] = batsman_runs.get(batsman,0) + runs

            if 'wicket' in ball:
                bowler_wickets[bowler] = bowler_wickets.get(bowler,0) + 1

        batsman_df = pd.DataFrame(
            batsman_runs.items(),
            columns=['Batsman','Runs']
        ).sort_values(by='Runs',ascending=False).head(4)

        bowler_df = pd.DataFrame(
            bowler_wickets.items(),
            columns=['Bowler','Wickets']
        ).sort_values(by='Wickets',ascending=False).head(4)

        scorecard = pd.concat([batsman_df, bowler_df], axis=1)

        scorecards.append(scorecard)

    return scorecards
```

```python
score1, score2 = get_scorecard(0)

print("First Innings Scorecard")
print(score1)

print("\nSecond Innings Scorecard")
print(score2)
```
